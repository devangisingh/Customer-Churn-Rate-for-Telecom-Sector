```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
tcc = pd.read_csv("./Telco-Customer-Churn.csv")
```

```python
tcc.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(tcc['MonthlyCharges'], kde=True, bins=30)
plt.title('Distribution of Monthly Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=tcc)
plt.title('Box Plot of Monthly Charges by Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
correlation = tcc.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(data=tcc, x='MonthlyCharges', y='TotalCharges', hue='Churn', style='Churn')
plt.title('Monthly Charges vs. Total Charges by Churn')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```
